A neural network that predicts the salary of an employee based on their number of years experience.

So, imagine if estimating an employees was an extremely simple. Base pay of 50k for someone with no experience and every additional year of experience gives you an additional 10k.

For the given problem statement, neural network is created that learns this relationship so that it would predict that an employee with 8 years of experience should earn 130k.

```python
#python_libraries
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
#salary_model
def salary_model(x_new):
    xs = [0,1,2,3,4,5,6] # number of year
    ys = [0.5,0.6,0.7,0.8,0.9,1.0,1.1] #salary 
    model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    model.compile(optimizer="sgd", loss='mean_squared_error')
    model.fit(xs, ys, epochs=500)
    return model.predict(x_new)[0][0]
```

```python
prediction = salary_model([8.0])
```

```python
print(prediction*100, " thousand dollars")
```
